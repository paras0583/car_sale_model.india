3. PLOTTING GRAPHS AND USEFULL INSIGHTS WE CAN SEE MARUTI SUZUKI VARIABLE HAS THE MOST COUNTS OF CARS.

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df,x='Brand',kde =True )
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

3.2 PLOTTING THE PIE CHART FOR COLUMN BRAND.üò§üò§üò§

```python
colors=['red','yellow','blue','green','grey']
labels = Brand_Labels.index
plt.figure(figsize=(15,15))
plt.pie(Brand_Labels,labels=labels,colors=colors)
plt.tight_layout()
plt.title('PIE CHART', fontsize=20)
plt.show()


```
