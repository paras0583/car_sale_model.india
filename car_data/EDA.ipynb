```python
fig , axs = plt.subplots(1,3, figsize=(15,4))
car_type_data = df['Car Type'].value_counts()
labels = car_type_data.index
axs[0].pie(car_type_data , labels = labels, colors=['red' , 'yellow' , 'blue' , 'green'], autopct='%1.1f%%')
plt.tight_layout()
axs[0].set_title("CAR TYPE", fontsize = 10)


Transmission_data = df['Transmission'].value_counts()
labels = Transmission_data.index
axs[1].pie(Transmission_data , labels = labels, colors=['red' , 'yellow' , 'blue' , 'green'], autopct='%1.1f%%')
plt.tight_layout()
axs[1].set_title("Transmission", fontsize = 10)


Fuel_data = df['Fuel Type'].value_counts()
labels = Fuel_data.index
axs[2].pie(Fuel_data , labels = labels, colors=['red' , 'yellow' , 'blue' , 'green'], autopct='%1.1f%%')
plt.tight_layout()
axs[2].set_title("Fuel", fontsize = 10)
plt.show()
```

3.6 NOW WE CAN ALSO COMPARE ACCIDENT WITH CAR BRAND üíÄ(I KNOW WHO GONNA TOP THIS LIST ALREADY SUZ...)

```python
accidental_brand = df[['Accidental', 'Brand']].copy()

# Add 3rd column with count of each Accidental type
accidental_brand['Accident Count'] = accidental_brand['Accidental'].map(df['Accidental'].value_counts())

# Filter only 'Yes' rows
yes_data = accidental_brand[accidental_brand['Accidental'].str.strip().str.lower() == 'yes']

print(yes_data.head())

```

3.7 DRAWING A PIECHAR FOR VISUALISATION WHICH BRAND OF CARS HOLD THE MOST NO OF ACCIDENTS TITLE ALTHOUGH I KNOW WHO GONNA TOP THE LIST AT LEAST FOR NOW üíÄüíÄ

```python
# Step 1: Filter only 'Yes' accidental cars
acc_yes = df[df['Accidental'].str.strip().str.lower() == 'yes']

# Step 2: Count of 'Yes' accidents brand-wise
brand_accident_counts = acc_yes['Brand'].value_counts()

# Step 3: Plot pie chart
plt.figure(figsize=(10, 10))
plt.pie(
    brand_accident_counts,
    labels=brand_accident_counts.index,
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Accidental Cars by Brand")
plt.axis('equal')  # For perfect circle
plt.tight_layout()
plt.show()
```
